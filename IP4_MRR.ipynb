**Part 11:** Create bar graphs for categorical variables.

```python
fig = plt.figure(figsize= (60,20))
GTD.country_txt.groupby(GTD.country_txt).count().plot(kind = 'bar', title = 'Acts of Terrorism by Country (1970 - 1997)', xlabel = 'Country Name', ylabel = 'Count', color = 'blue')
plt.savefig('Terrorist Attacks by Country.png')
```

```python
fig = plt.figure(figsize= (20,20))
GTD.targtype1_txt.groupby(GTD.targtype1_txt).count().plot(kind = 'bar', title = 'Target of the Terrorist Attacks (1970 - 1997)', xlabel = 'Target of Attack', ylabel = 'Count')
plt.savefig('Targets of Terrorist Attacks.png')
```

```python
fig = plt.figure(figsize= (60,20))
GTD.natlty1_txt.groupby(GTD.natlty1_txt).count().plot(kind = 'bar', title ='Nationality of the Target of Terroist Attack (1970 - 1997)', xlabel = 'Nationality', ylabel = 'Count', color = 'purple')
plt.savefig('Nationality of Terrorist Attack Target.png')
```

```python
fig = plt.figure(figsize= (20,20))
GTD.attacktype1_txt.groupby(GTD.attacktype1_txt).count().plot(kind = 'bar', title = 'Type of Terrorist Attack (1970 - 1997)', xlabel = 'Type of Attack', ylabel = 'Count', color = 'blue')
plt.savefig('Type of Terrorist Attack.png')
```

```python
fig = plt.figure(figsize = (20,10))
United_Kingdom = GTD.query("country_txt == 'United Kingdom'")
United_Kingdom.iyear.groupby(United_Kingdom.iyear).count().plot(kind = 'bar', title = 'Terrorist Attacks per Year in United Kingdom', xlabel = 'Years')
```

```python
fig = plt.figure(figsize = (12,7))
fig.subplots_adjust(wspace = 0.5, hspace = 0.4)
ax1 = fig.add_subplot(221)
United_States.iyear.groupby(United_States.iyear).count().plot(kind = 'bar', title = 'Terrorist Attacks per Year in the United States', xlabel = 'Years', ylabel = 'Number of Attacks', color = 'purple')
ax2 = fig.add_subplot(222)
Iran.iyear.groupby(Iran.iyear).count().plot(kind = 'bar', title = 'Terrorist Attacks per Year in Iran', xlabel = 'Years')
ax3 = fig.add_subplot(223)
El_Salvador.iyear.groupby(El_Salvador.iyear).count().plot(kind = 'bar', title = 'Terrorist Attacks per Year in El Salvador', xlabel = 'Years')
ax4 = fig.add_subplot(224)
United_Kingdom.iyear.groupby(United_Kingdom.iyear).count().plot(kind = 'bar', title = 'Terrorist Attacks per Year in United Kingdom', xlabel = 'Years')
plt.savefig('Attacks by Country Across the Years.png')
```
